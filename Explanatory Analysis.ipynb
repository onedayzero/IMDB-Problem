# Setup

```python
import pandas as pd
import numpy as np
```

```python
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000.head()
```

```python
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2001.head()
```

```python
# Create 1 df from both dfs
df = pd.concat([df_2000,df_2001])
df.head()
```

# Analysis

```python
finance_filter = df[(df['budget'] > 0) & (df['revenue'] > 0)]
finance_filter.count()
```

## How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

There appears to be 349 results that have their budget or revenue data.

```python
df = finance_filter
df.head()
```

```python
# Count of movies in each category
df.certification.value_counts()
```

```python
# mean revenue by category
cert_filter = df.groupby(['certification']).mean()
cert_filter.revenue.round()
```

```python
# mean budget by category
cert_filter = df.groupby(['certification']).mean()
cert_filter.budget.round()
```
